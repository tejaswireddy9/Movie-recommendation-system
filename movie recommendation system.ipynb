```python
import ast
def extract(teju):
    l=[]
    for j in teju:
        li=[]
        for i in ast.literal_eval(j):
             li.append(i["name"])
        l.append(li)
    return l
```

```python
def extract_director(teju):
        l=[]
        for j in teju:
            li=[]
            for i in ast.literal_eval(j):
                if(i.get("job")=="Director"):
                    li.append(i.get("name")) 
                    break
            l.append(li)
        return l
```

```python
woo=extract_director(movies["crew"])
woo
```

```python
movies["crew"]=woo
```

```python
movies
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["genres"]
```

```python
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies["overview"]
```

```python
movies["tags"]=movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]+movies["overview"]
```

```python
df=movies[["id","original_title","tags"]]
df
```

```python
df["tags"]=df["tags"].apply(lambda x:[i.lower() for i in x])
```

```python
df["tags"]
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df.head()
```

```python
df["tags"].shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
co_ve = CountVectorizer(max_features=4000,stop_words='english')
```

```python
vr = co_ve.fit_transform(df['tags']).toarray()
vr
```

```python
co_ve.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
post=PorterStemmer()
```

```python
def stemming(akshi):
    l1=[]
    for i in akshi.split():
        a=post.stem(i)
        l1.append(a)
    return " ".join(l1)
```

```python
df["tags"]=df["tags"].apply(stemming)
df["tags"]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosim = cosine_similarity(vr)
cosim
```

```python
cosim.shape
```

```python
def recommendation(movie_name):
    i = df[df['original_title'] == movie_name].index[0]
    cdis = sorted(list(enumerate(cosim[i])),reverse=True,key = lambda y: y[1])
    for j in cdis[1:6]:
        print(df.iloc[j[0]].original_title)
```

```python
recommendation("Batman Begins")
```
